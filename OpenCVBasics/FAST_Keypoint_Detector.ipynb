# FAST Keypoint Detector

This descriptor is used to detect corners in a image. The algorithm is simple and very fast in execution and hence is a popular method to detect corners. 

## How it works

For every pixel ***p*** the algorithm looks for ***n*** pixels along a circular perimeter of radius ***r*** about the pixel which are either brighter or darker than the center pixel and identifies the pixel as one having a corner if the condition satisfies. 

![FAST Explained](image/fast_raw_example-865x412.jpg)

In this example, the perimeter of the circle consists of 16 pixels which corresponds to a radius of 3 px. It is common practise to use a radius of 3 and check for ***n*** = 9 or 12

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

image = cv2.imread('image/next.png')
orig = image.copy()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Using cv2's FastFeatureDetector to detect keypoints
detector = cv2.FastFeatureDetector_create()
kps = detector.detect(gray, None)

# loop over the keypoints and draw them
for kp in kps:
    r = int(0.5 * kp.size)
    (x, y) = np.int0(kp.pt)
    cv2.circle(image, (x, y), r, (0, 255, 255), 2)
    
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```
